## Student Performance Analyzer

2. Student Performance Analyzer
Create a Student Performance Analyzer that stores academic records and derives
meaningful insights.
The system should allow adding students with roll numbers and names, entering marks for
multiple subjects, and storing them in structured formats. It should calculate individual
averages, subject-wise toppers, pass/fail status based on defined rules, and overall ranking
without using built-in sorting functions.
Marks must be validated, grading rules should differ across subjects, and summary reports
must be generated using loops and conditional logic. The program must be fully
menu-driven and modular.

```python
students = {}


def add_student():
    roll = input("Enter Roll No: ")
    name = input("Enter Name: ")

    students[roll] = {
        "name": name,
        "math": 0,
        "science": 0,
        "english": 0
    }

    print("Student Added!")


def enter_marks():
    roll = input("Enter Roll No: ")

    if roll not in students:
        print("Student not found!")
        return

    students[roll]["math"] = int(input("Math Marks: "))
    students[roll]["science"] = int(input("Science Marks: "))
    students[roll]["english"] = int(input("English Marks: "))

    print("Marks Saved!")


def average(roll):
    s = students[roll]
    return (s["math"] + s["science"] + s["english"]) / 3


def status(roll):
    s = students[roll]

    if s["math"] >= 40 and s["science"] >= 35 and s["english"] >= 30:
        return "PASS"
    else:
        return "FAIL"


def report():
    roll = input("Enter Roll No: ")

    if roll not in students:
        print("Student not found!")
        return

    s = students[roll]

    print("\nName:", s["name"])
    print("Math:", s["math"])
    print("Science:", s["science"])
    print("English:", s["english"])
    print("Average:", average(roll))
    print("Status:", status(roll))


def topper():
    highest = -1
    top_name = ""

    for roll in students:
        if students[roll]["math"] > highest:
            highest = students[roll]["math"]
            top_name = students[roll]["name"]

    print("\nMath Topper:", top_name, "-", highest)


def ranking():
    data = []

    for roll in students:
        data.append([students[roll]["name"], average(roll)])

    n = len(data)

    for i in range(n):
        for j in range(n - i - 1):
            if data[j][1] < data[j + 1][1]:
                data[j], data[j + 1] = data[j + 1], data[j]

    print("\nRanking")
    for i in range(len(data)):
        print(i + 1, ".", data[i][0], "-", round(data[i][1], 2))


while True:
    print("\n--- Student Performance Analyzer ---")
    print("1. Add Student")
    print("2. Enter Marks")
    print("3. Student Report")
    print("4. Topper")
    print("5. Ranking")
    print("6. Exit")

    choice = input("Enter Choice: ")

    if choice == "1":
        add_student()

    elif choice == "2":
        enter_marks()

    elif choice == "3":
        report()

    elif choice == "4":
        topper()

    elif choice == "5":
        ranking()

    elif choice == "6":
        break

    else:
        print("Invalid Choice")
```
